# 获取输入 

目前我们无法对窗口直接做任何事 

* 添加处理输入的函数
* 添加调用处理输入的主循环函数
* 记录游戏状态 供主循环使用 

```c++
%%file ../code/2_input/MainGame.h
#pragma once
#include <SDL.h>
#include <GL/glew.h>

enum class GameState {PLAY, EXIT};

class MainGame {
public:
    MainGame();
    ~MainGame();
    // 运行游戏 
    void run();
private:
    void initSystems();      // -- update
    // 处理输入 
    void processInput();     // -- new
    // 主游戏循环 
    void gameLoop();         // -- new
    SDL_Window *_window;
    int _screenWidth; 
    int _screenHeight;
    // 指示游戏状态 
    GameState _gameState;    // -- new
};
```

## SDL_PollEvent

轮询当前未决的事件。如果事件队列存在事件，则绑定到传入事件指针并返回1；否则事件指针为NULL，且返回0

```c++
%%file ../code/2_input/MainGame.cpp

#include "MainGame.h"
#include <iostream>

MainGame::MainGame() { 
    _window = nullptr;
    _screenWidth = 1024;
    _screenHeight = 768;
    _gameState = GameState::PLAY;
}

MainGame::~MainGame() {
}

void MainGame::run() {
    initSystems();
    gameLoop();
}

void MainGame::initSystems() {
    SDL_Init(SDL_INIT_EVERYTHING);

    _window = SDL_CreateWindow("Game Engine", SDL_WINDOWPOS_CENTERED, 
                                SDL_WINDOWPOS_CENTERED, _screenWidth, _screenHeight,
                                SDL_WINDOW_OPENGL);
    
    SDL_GL_CreateContext(_window);
    SDL_GL_SwapWindow(_window);
}

// 处理输入 
void MainGame::processInput() {
    SDL_Event evnt;
    // 返回一个 整数 1: 存在事件 0: 没有事件 
    // 传入一个 SDL_Event * 
    // 如果存在事件 会将 事件绑定在 传入 指针 
    while(SDL_PollEvent(&evnt)) {
        switch (evnt.type) {
            // 搜索 SDL_Event 查看更多事件 以及其详细用法
            case SDL_QUIT:
                _gameState = GameState::EXIT;
                break;
            case SDL_MOUSEMOTION:
                // 要使用的属性 在 wiki 中 SDL_Event Field 说明
                std::cout << evnt.motion.x << " " << evnt.motion.y << std::endl;
                break;
        }
    }
}
    
// 主游戏循环 其中重复处理输入 除非 _gameState 被设为 EXIT
void MainGame::gameLoop() {
    while (_gameState != GameState::EXIT) {
        processInput();
    }
}
```

### 现在可以去掉 main.cpp 中维持 窗口的部分

```c++
%%file ../code/2_input/main.cpp

#include <iostream>
#include "MainGame.h"

int main(int argc, char **arg) {
    MainGame mainGame;
    mainGame.run();
    return 0;
}
```
